```python
# Function to analyze post engagement
def analyze_engagement(likes):
    total_likes = 0

    # Calculate total likes
    for like in likes:
        total_likes += like

    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Likes received over different days
likes = [250, 300, 200, 300]

# Call function
analyze_engagement(likes)
```

```python
# Function to check medicine stock
def check_stock(stock):
    print("Medicine Stock:", stock)

    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


# Example stock quantity
medicine_stock = 6

# Call function
check_stock(medicine_stock)
```

```python
# Function to check rainfall adequacy
def check_rainfall(rainfall_data, required_level):
    total_rainfall = 0

    # Calculate total rainfall
    for rainfall in rainfall_data:
        total_rainfall += rainfall

    average_rainfall = total_rainfall / len(rainfall_data)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average_rainfall)
    print("Rainfall Status:", status)


# Seasonal rainfall data (in mm)
rainfall_data = [70, 75, 68, 80, 67]

# Required rainfall level
required_level = 70

# Call function
check_rainfall(rainfall_data, required_level)
```

```python
# Function to detect duplicate usernames
def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# List of usernames
usernames = ["ravi123", "anita45", "ravi123", "kiran99"]

# Call function
check_duplicates(usernames)
```

```python
# Function to filter premium crop prices
def filter_premium_crops(crop_prices):
    premium_crops = []

    for price in crop_prices:
        if price > 2000:
            premium_crops.append(price)

    return premium_crops


# Crop prices
crop_prices = [1800, 2500, 1500, 3200, 1900]

# Get premium crops
premium = filter_premium_crops(crop_prices)

print("Premium Crops:", premium)
```

```python
# Function to check application health
def check_application_health(error_count):
    print("Error Count:", error_count)

    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("System Status:", status)


# Example error count
errors = 7

# Call function
check_application_health(errors)
```

```python
# Function to check transaction limit
def check_transaction(amount):
    daily_limit = 50000

    print("Transaction Amount:", amount)

    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example transaction amount
transaction_amount = 60000

# Call function
check_transaction(transaction_amount)
```

```python
# Function to check exam eligibility
def check_exam_eligibility(attendance):
    present_days = 0

    # Count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    attendance_percentage = (present_days / len(attendance)) * 100

    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)


# Attendance record
attendance = ["P", "P", "A", "P", "P"]

# Call function
check_exam_eligibility(attendance)
```

```python
# Function to calculate electricity bill
def calculate_bill(units):
    bill = 0

    # First 100 units
    if units <= 100:
        bill = units * 3

    # Next 100 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    # Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example input
units_consumed = 250

bill, status = calculate_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Electricity Bill: ₹", bill)
print("Usage Status:", status)
```
