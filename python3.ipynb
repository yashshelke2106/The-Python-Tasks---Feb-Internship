```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest score
highest_score = max(employees.values())

# Find all employees with the highest score
top_performers = []

for employee, score in employees.items():
    if score == highest_score:
        top_performers.append(employee)

print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {highest_score})")
```

```python
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove common punctuation
for char in ".,!?;:":
    query = query.replace(char, "")

# Split into words
words = query.split()

# Count frequency
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Display words occurring more than once
repeated_keywords = {}

for word, count in frequency.items():
    if count > 1:
        repeated_keywords[word] = count

print(repeated_keywords)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, reading in enumerate(sensor_readings):
    if reading % 2 == 0:
        valid_readings.append((hour, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domain occurrences
for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Calculate percentages
total_users = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Calculate spike threshold (30% above average)
threshold = average_sales * 1.3

# Detect sales spikes
for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
```

```python
# List of registered user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency of each user ID
user_count = {}

# Count occurrences
for user_id in user_ids:
    if user_id in user_count:
        user_count[user_id] += 1
    else:
        user_count[user_id] = 1

# Display duplicates only
for user_id, count in user_count.items():
    if count > 1:
        print(f"{user_id} → {count} times")
        
```
