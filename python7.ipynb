```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.is_on = False

    def turn_on(self):
        self.is_on = True
        print(f"{self.light_name} is ON")

    def turn_off(self):
        self.is_on = False
        print(f"{self.light_name} is OFF")


# Main Program
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ").upper()

light = SmartLight(light_name)

if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid Action")
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# Create object
employee = Employee(name, emp_id, department)

# Display ID card
employee.display_id_card()
```

```python
# Class to store contact details
class Contact:
    
    # Constructor
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Creating object
contact1 = Contact("Anita", 9876543210)

# Display contact details
contact1.display_contact()
```

```python
# Class to store product details
class Product:

    # Constructor
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    # Method to display price tag
    def display_price_tag(self):
        print("Product:", self.product_name)
        print("Price: ₹{}".format(self.price))


# Creating object
product1 = Product("Headphones", 2499)

# Display product details
product1.display_price_tag()
```

```python
# Class to store movie details
class Movie:

    # Constructor
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    # Method to display movie details
    def display_movie(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")


# Creating object
movie1 = Movie("Inception", 4.8)

# Display movie details
movie1.display_movie()
```

```python
# Class to store delivery details
class Delivery:

    # Constructor
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to display delivery details
    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Creating object
delivery1 = Delivery("Suman", "Hyderabad")

# Display delivery details
delivery1.display_details()
```
