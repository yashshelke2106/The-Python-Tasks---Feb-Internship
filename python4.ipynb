```python
# Function to validate recharge

def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    if plan not in valid_plans:
        print("Invalid recharge plan.")
        return False

    return True


# Retry until valid input is provided
while True:
    amount = int(input("Enter recharge amount: ₹"))
    plan = int(input("Enter recharge plan: ₹"))

    if validate_recharge(amount, plan):
        print("Recharge Successful!")
        break
    else:
        print("Please try again.\n")
```

```python
# Function to check stock status
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


# Product inventory
products = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 15,
    "Monitor": 8
}

# Call function
check_inventory(products)
```

```python
# Function to process student result
def process_result(marks):
    total = 0

    # Calculate total marks
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average Marks: {average:.2f}\nResult: Pass"
    else:
        return f"Average Marks: {average:.2f}\nResult: Fail"


# Student marks
marks = [60, 55, 70, 45]

# Display result
print(process_result(marks))
```

```python
# Function to calculate cab fare
def calculate_fare(distance, peak_hour):
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak_hour.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare


# Retry option
while True:
    distance = float(input("Enter distance (km): "))
    peak_hour = input("Peak Hour? (Yes/No): ")

    total_fare = calculate_fare(distance, peak_hour)

    print(f"Total Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (Yes/No): ")

    if retry.lower() != "yes":
        print("Thank you for using the Cab Fare Estimator.")
        break
```

```python
# Function to check attendance eligibility
def check_eligibility(attendance):
    present_days = 0

    # Count present days
    for status in attendance:
        if status == "P":
            present_days += 1

    attendance_percentage = (present_days / len(attendance)) * 100

    if attendance_percentage >= 75:
        return f"Attendance Percentage: {attendance_percentage:.2f}%\nStatus: Eligible"
    else:
        return f"Attendance Percentage: {attendance_percentage:.2f}%\nStatus: Not Eligible"


# Attendance record
attendance = ["P", "P", "A", "P", "P"]

# Display result
print(check_eligibility(attendance))
```

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False

    # Check password length
    if len(password) < 8:
        return "Weak Password: Minimum length should be 8 characters."

    # Check for digit and special character
    for char in password:
        if char.isdigit():
            has_digit = True

        if char in "@#$":
            has_special = True

    # Validate password
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password: Must contain at least one digit and one special character."


# User input
password = input("Enter Password: ")

print(check_password(password))
```
