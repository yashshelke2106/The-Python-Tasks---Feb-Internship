```python
sentence = input("Enter sentence: ")

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
employees = {
    "Ravi": 79000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num

    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

```python
text = "pythonp"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```
