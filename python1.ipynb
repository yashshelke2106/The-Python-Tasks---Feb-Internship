```python
username = "admin"
password = "1234"

# User input
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Check credentials
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Pass Students:", pass_count)
print("Fail Students:", fail_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())

print("Cleaned Names:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}'")
    print(f"Length: {length}")

    if length > 10:
        print("Flag: Long message")
    else:
        print("Flag: Normal message")

    print()
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total Error Count:", error_count)
```
