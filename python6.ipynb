```python
# Function to manage parking lot
def parking_management(capacity, vehicle_logs):
    current_vehicles = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)

    if current_vehicles > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Function Call
parking_management(capacity, vehicle_logs)
```

```python
# Function to estimate delivery time
def estimate_delivery_time(distance, traffic, weather):
    # Base delivery time
    eta = distance * 5

    # Traffic delays
    if traffic.lower() == "high":
        eta += 10
    elif traffic.lower() == "medium":
        eta += 5

    # Weather delays
    if weather.lower() == "rainy":
        eta += 5
    elif weather.lower() == "stormy":
        eta += 15

    return eta


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

# Function Call
delivery_time = estimate_delivery_time(distance, traffic, weather)

print("Estimated Delivery Time:", delivery_time, "minutes")
```

```python
# Function to analyze seat occupancy
def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", occupancy, "%")
    print("Show Status:", status)

    if occupancy >= 90:
        print("Suggestion: Open Additional Shows")


# Sample Input
total_seats = 200
booked_seats = [1] * 150  # 150 booked seats

# Function Call
analyze_occupancy(total_seats, booked_seats)
```

```python
# Function to classify server load
def classify_server_load(cpu_readings):
    total = 0

    # Calculate total CPU usage
    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)

    # Classify server status
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", average_cpu, "%")
    print("Server Status:", status)


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

# Function Call
classify_server_load(cpu_readings)
```

```python
# Function to monitor classroom resources
def monitor_resources(resource_usage):
    overused_resources = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused_resources.append(resource)

    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function Call
monitor_resources(resource_usage)
```

```python
# Function to manage event registrations
def manage_registrations(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # Process each registration
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Determine registration status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_capacity = 100
registrations = 105

# Function Call
manage_registrations(event_capacity, registrations)
```
